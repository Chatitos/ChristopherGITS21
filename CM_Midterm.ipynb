```python
city = ['Seattle' , 'San Francisco' , 'Los Angeles' , 'Denver' , 'Salt Lake City' , 'Phoenix' , 'Chicago' , 'St. Louis' , 'Dallas' , 'Boston' , 'Washington D.C' , 'Miami']
totaldays = [306,314,275,366,366,366,366,366,306,306,366,275]
poordays = [14,22,119,34,21,70,28,10,24,0,3,2]
aqimed = [42,45,81,58,50,75,56,52,49,42,42,42]
zone = ['west','west', 'west','mountain','mountain','mountain','central','central','central','east','east','east']
```

```python
import pandas as pd

aqistats = pd.DataFrame({'City':city,'Total_Days':totaldays,'Poor_Days':poordays,'AQI_Median':aqimed,'Zone':zone})
```

```python
city10 = ['Seattle' , 'San Francisco' , 'Los Angeles' , 'Denver' , 'Salt Lake City' , 'Phoenix' , 'Chicago' , 'St. Louis' , 'Dallas' , 'Boston' , 'Washington D.C' , 'Miami', 'Springfield','Providence','Appleton','Racine','Boulder','Fort Collins','Aberdeen','Bellingham']
totaldays10 = [365,365,365,365,365,365,365,365,365,365,365,365,365,365,261,192,365,365,365,355]
poordays10 = [3,14,115,38,24,59,43,103,28,10,50,6,12,20,4,6,12,21,0,0]
aqimed10 = [41,48,77,61,51,74,71,74,52,47,55,48,43,44,38,42,45,51,16,27]
zone = ['west','west','west','mountain','mountain','mountain','central','central','central','east','east','east','east','east','central','central','mountain','mountain','west','west']
```

```python
allcities['Poorday_Rate'] = allcities['Poor_Days']/allcities['scale']
```

```python
allcities['Poorday_Rate'] = allcities['Poorday_Rate'].astype(int)
```
